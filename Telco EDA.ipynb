# Feature Engineering

```python
df1 = df
```

```python
df1.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
df1.isnull().sum()
```

```python
df1.loc[df1['TotalCharges'].isnull() == True]
```

Observation : After converting the categorical column Total charges into numerical we found 11 Nan values which is very less 0.15% of 
    total records. So, it is safe to ignore them for further processing

```python
df1.dropna(how = 'any', inplace = True)
```

```python
df1.isnull().sum()
```

Making new column from tenure 

```python
df1.tenure.max()
```

```python
labels = ["{0} - {1}".format(i, i+11) for i in range(1, 72, 12)]
# labels
df1['tenure_group'] = pd.cut(df.tenure, range(1, 80, 12), right = False, labels = labels)
```

```python
df1['tenure_group'].value_counts()
```

Remove the columns which are not required for the processing

```python
df1.drop(columns = ['customerID', 'tenure'], axis = 1, inplace =True)
df1.head()
```

```python
df1.shape
```

# Data Exploration

```python
for i in df1:
    plt.figure(figsize = (16,9))
    sns.countplot(data = df1, x=df1[i], hue = 'Churn')
```

Converting the target variable Churn in a binary numeric variable

```python
df1['Churn'] = np.where(df1.Churn == 'Yes', 1,0)
```

# One Hot Encoding

```python
df_dummies = pd.get_dummies(df1)
df_dummies.head()
```

Relation between monthly charges and total chargers

```python
sns.lmplot(x = 'MonthlyCharges', y = 'TotalCharges', fit_reg = False, data = df_dummies)
```

```python
plt.figure(figsize = (16,9))
df_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```

# Bivariate Analysis

```python
new_df_target0 = df1.loc[df1['Churn'] == 0]
new_df_target1 = df1.loc[df1['Churn'] == 1]
```

```python
def uniplot(df, col, title, hue = None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams['axes.labelsize'] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 32
        
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4 * len(temp.unique())
    fig.set_size_inches(width, 8)
    plt.xticks(rotation = 90)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x = col, order = df[col].value_counts().index, hue = hue, palette='bright')
```

```python
uniplot(new_df_target0, 'Partner', title = 'Distribution of gender for non churned customers', hue = 'gender')
```

```python
uniplot(new_df_target1, 'Partner', title = 'Distribution of gender for churned customers', hue = 'gender')
```

```python
uniplot(new_df_target1, 'PaymentMethod', title = 'Distribution of gender for churned customers', hue = 'gender')
```
